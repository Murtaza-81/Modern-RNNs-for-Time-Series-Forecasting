```python
import torch
import pandas as pd
from dataset import load_data
from models.baseline import GRUModel
from models.igru_variant import iGRUModel
from utils import rmse, mae, mape

_, _, test_loader, dim = load_data()

models = {
    "GRU": GRUModel(dim),
    "iGRU": iGRUModel(dim)
}

results = []

for name, model in models.items():
    model.load_state_dict(torch.load(f"{name}.pt"))

    ys, preds = [], []

    with torch.no_grad():
        for x, y in test_loader:
            p = model(x)
            ys.append(y)
            preds.append(p)

    y = torch.cat(ys)
    yhat = torch.cat(preds)

    res = {
        "Model": name,
        "RMSE": rmse(yhat, y),
        "MAE": mae(yhat, y),
        "MAPE": mape(yhat, y)
    }

    print(res)
    results.append(res)

pd.DataFrame(results).to_csv("results/results.csv", index=False)
```
